3.	What is the distribution of the winning percentage for all teams? Which teams are in the top 5 in terms of winning percentage? (Winning percentage= (GamesWon / GamesPlayed)*100)

```python
df['GamesWon']=df['GamesWon'].astype(float)
df['GamesPlayed']=df['GamesPlayed'].astype(float)
df['Winning percentage']=(df['GamesWon']/df['GamesPlayed'])*100
df
```

4.	Is there a significant difference in the winning percentage for teams which have attained a best position between 1-3 and those teams which have had a best position between 4-7.

```python
df['BestPosition']=df['BestPosition'].astype(int)
a=df[(df['BestPosition']>=1)&(df['BestPosition']<=3)]
b=df[(df['BestPosition']>=4)&(df['BestPosition']<=7)]
amean=a['Winning percentage'].mean()
bmean=b['Winning percentage'].mean()
if(amean!=bmean):
    print('there is a significant difference of',amean-bmean)
else:
    print('no difference')
```

b)	Estimate a 95% confidence interval for the true mean blood pressure for all 60-year old women with glaucoma.

```python
import scipy.stats as stats
left_interval,right_interval=stats.norm.interval(0.95,u,stand_error)
print(left_interval,right_interval)
```

```python
Z=stats.norm.isf(q=0.05/2)
xbar=u
sigma=S
Lower_Interval = xbar - (Z * (sigma / np.sqrt(n)))
Upper_Interval = xbar + (Z * (sigma / np.sqrt(n)))
print('90 %s confidence interval for population mean is %1.4f  to %1.4f' % ('%', Lower_Interval , Upper_Interval))
```

ANS:since the distance is normally distributed and the 2 samples are independent of each other it is best to go for a unpaired t test,so that we can say whether current golf ball is performing good or it is the new one 

```python
df=pd.read_csv('Golf.csv')
df
```

2.	Analyze the data to provide the hypothesis testing conclusion. What is the p-value for your test? What is your recommendation for Par Inc.?

```python
alpha=0.05
curr=df['Current']
new=df['New']
t,p=stats.ttest_ind(curr,new)
print(p)
if(p>alpha):
    print('Acceept H0,no difference')
else:
    print('Reject H0,there is different')
```
